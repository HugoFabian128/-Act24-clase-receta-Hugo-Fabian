```python
class Receta:
    def __init__(self, nombre, tiempo, dificultad, porciones):
        self.nombre = nombre
        self.tiempo_preparacion = tiempo
        self.dificultad = dificultad
        self.porciones = porciones

    def listar_ingredientes(self):
        ingredientes = input(f"Ingresa los ingredientes para {self.nombre} (separados por coma): ")
        print(f"\n> Lista de insumos: {ingredientes}")

    def mezclar(self):
        print(f"> Acción: Mezclando con cuidado para {self.porciones} porciones.")

    def hornear(self):
        print(f"> Acción: Horneando... ¡Listo en {self.tiempo_preparacion} minutos!")

def crear_receta(numero):
    print(f"\n--- DATOS DE LA RECETA {numero} ---")
    nom = input("Nombre de la receta: ")
    tie = input("Tiempo de preparación (min): ")
    dif = input("Dificultad (Baja/Media/Alta): ")
    por = input("Cantidad de porciones: ")

    # Crear objeto
    objeto = Receta(nom, tie, dif, por)

    # Ejecutar funciones
    objeto.listar_ingredientes()
    objeto.mezclar()
    objeto.hornear()
    return objeto

# Ejecución
receta1 = crear_receta(1)
receta2 = crear_receta(2)

print("\n--- PROCESO FINALIZADO ---")
print("Programa realizado por Hugo Fabian NL = 15")
```
